# Python Basics for Reading Course Notebooks

This notebook is a short, guided practice notebook. It assumes you already completed steps 1-4 in `lab0_01_llm_foundations/01_instructions.md` so your virtual environment, packages, and Jupyter are ready.

The goal is not to write complex programs. The goal is to practice reading the same Python patterns that appear in the later labs.

```python
import csv
import json
from pathlib import Path

LAB_NAME = "lab0_00_python_basics"
lab_dir = Path.cwd()
if lab_dir.name != LAB_NAME:
    raise RuntimeError(
        f"Open this notebook from the {LAB_NAME} folder after completing steps 1-4 in lab0_01_llm_foundations."
    )

data_dir = lab_dir / "data"
print("Notebook folder:", lab_dir.name)
print("Data files:", [path.name for path in sorted(data_dir.iterdir())])
```

## 1. Variables and Simple Values

Variables store values. In later labs, you will see variables that store case IDs, prompts, model outputs, file paths, and rows loaded from evidence files.

```python
case_id = "PY-INTRO-001"
question = "Did the employee send a file to an external contact?"
confidence = 0.4

print("Case ID:", case_id)
print("Question:", question)
print("Confidence:", confidence)
```

## 2. Lists and Dictionaries

Lists store ordered items. Dictionaries store key-value pairs. These two structures show up everywhere in the course notebooks.

```python
artifacts = ["timeline.csv", "messages.csv", "chain_of_custody.csv"]

response = {
    "model": "tiny-demo",
    "seconds": 1.2,
    "summary": "Practice output"
}

case_packet = {
    "case_id": case_id,
    "question": question,
    "artifacts": artifacts
}

print("First artifact:", artifacts[0])
print("Model name:", response["model"])
print("Question from case packet:", case_packet["question"])
```

## 3. `if` and `for`

An `if` statement checks a condition. A `for` loop repeats a small block of code one item at a time.

```python
toolbox = {
    "read_manifest": "Read the case manifest",
    "read_timeline": "Read timeline rows",
    "draft_summary": "Write a short case summary"
}

if confidence < 0.5:
    print("Confidence is still low.")

print("\nTools:")
for name in toolbox:
    print(f"- {name}: {toolbox[name]}")
```

## 4. Functions and Returned Dictionaries

Later notebooks often call helper functions that return dictionaries. You do not need to memorize every line. Focus on the input, the return value, and how the returned keys are used.

```python
def summarize_case(packet: dict) -> dict:
    return {
        "case_id": packet["case_id"],
        "artifact_count": len(packet["artifacts"]),
        "question": packet["question"]
    }

summary = summarize_case(case_packet)
print("Summary dictionary:", summary)
print("Artifact count:", summary["artifact_count"])
```

## 5. Reading JSON and CSV Data

Many later labs load case artifacts from JSON and CSV files. JSON usually turns into dictionaries and lists. CSV often turns into a list of row dictionaries.

```python
with (data_dir / "case_packet.json").open("r", encoding="utf-8") as handle:
    loaded_packet = json.load(handle)

with (data_dir / "timeline.csv").open("r", encoding="utf-8", newline="") as handle:
    timeline_rows = list(csv.DictReader(handle))

print("Loaded case ID:", loaded_packet["case_id"])
print("First note:", loaded_packet["notes"][0])
print("First timestamp:", timeline_rows[0]["timestamp"])
print("First event:", timeline_rows[0]["event"])
```

## 6. Mini Practice With Later-Lab Patterns

This final section combines the same small patterns you will see later: helper functions, returned dictionaries, looping through a toolbox, and reading fields from loaded data.

If you want a tiny extra challenge, change `selected_case_key` or `selected_row_key` before you rerun the cell.

```python
def read_case_packet(path: Path) -> dict:
    with path.open("r", encoding="utf-8") as handle:
        return json.load(handle)


def read_timeline(path: Path) -> list[dict]:
    with path.open("r", encoding="utf-8", newline="") as handle:
        return list(csv.DictReader(handle))


def build_toolbox() -> dict:
    return {
        "read_manifest": "Read high-level case details",
        "read_timeline": "Review timeline rows",
        "draft_summary": "Write a short practice summary"
    }


packet = read_case_packet(data_dir / "case_packet.json")
rows = read_timeline(data_dir / "timeline.csv")
practice_toolbox = build_toolbox()

practice_result = {
    "case_id": packet["case_id"],
    "question": packet["question"],
    "first_timestamp": rows[0]["timestamp"],
    "first_event": rows[0]["event"],
    "tool_count": len(practice_toolbox),
    "model_name": response["model"]
}

print("Practice result:", practice_result)

print("\nTool names:")
for name in practice_toolbox:
    print("-", name)

selected_case_key = "device_owner"
selected_row_key = "source"
print("\nSelected case value:", packet[selected_case_key])
print("Selected row value:", rows[0][selected_row_key])
```
